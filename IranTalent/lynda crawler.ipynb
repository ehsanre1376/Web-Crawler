```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
from selenium import webdriver
import time

number = int(input("please enter number of course you need -> "))

print("please waite...!")
print("clicking the show more button...")

options = webdriver.ChromeOptions()
options.add_argument('headless')

driver = webdriver.Chrome(executable_path=r"D:\faradars\files\lesson8\chromedriver.exe", options=options)
driver.get("https://www.lynda.com/Design-training-tutorials/40-0.html")

show_more_button = driver.find_element_by_css_selector("#category-courses > li.show-more-row")
n = len(driver.find_elements_by_css_selector("#category-courses > li.card-cont "))
print(n)

time.sleep(10)

if number>50:
    show_more_button.click()
    
    while True:
        temp1 = driver.find_element_by_css_selector("#category-courses > li.show-more-row")
        temp2 = driver.find_element_by_css_selector("#category-courses > li:nth-of-type("+str(n+1)+")")
        if temp1!=temp2:
            n = len(driver.find_elements_by_css_selector("#category-courses > li.card-cont "))
            print(n)
            if n>=number or temp1.is_displayed()==False:
                break
            temp1.click()
#######################################################################################

soup = BeautifulSoup(driver.page_source, 'html.parser')
# courses = soup.select("#category-courses > li.card-cont ")
# print(len(courses))

print("crawling category page...")

#link
temp = soup.select("#category-courses > li > div > div > div.col-xs-8.col-sm-9.card-meta-data > a")[:number]
link = [i['href'] for i in temp]

#title
temp = soup.select("#category-courses > li > div > div > div.col-xs-8.col-sm-9.card-meta-data > a > div.title-author-info > h3")[:number]
title = [i.text for i in temp]

#short_description
temp = soup.select("#category-courses > li > div > div > div.col-xs-8.col-sm-9.card-meta-data > a > div.title-author-info > div")[:number]
short_description = [i.text for i in temp]

#duration
temp = soup.select("#category-courses > li > div > div > div.col-xs-8.col-sm-9.card-meta-data > a > div.meta > span.meta-duration")[:number]
duration = [i.text for i in temp]

#skill_level
temp = soup.select("#category-courses > li > div > div > div.col-xs-8.col-sm-9.card-meta-data > a > div.meta > span.meta-level")[:number]
skill_level = [i.text for i in temp]

#release_date
temp = soup.select("#category-courses > li > div > div > div.col-xs-8.col-sm-9.card-meta-data > a > div.meta > span.meta-released > span")[:number]
release_date = [i.text for i in temp]

print("crawling courses page...")
description, topics, author_name, author_img, tags, main_img = [], [], [], [], [], []

for link1 in tqdm(link):
    try:
        site = requests.get(link1)
    except:
        time.sleep(5)
        site = requests.get(link1)
        
    soup = BeautifulSoup(site.text, "html.parser")

    #description
    temp = soup.select("#overview > div > div.col-sm-7.col-md-6.col-lg-8.course-description > div")
    description.append(temp[0].find(text = True).strip())

    #topics
    temp = soup.select("#overview > div > div.col-sm-7.col-md-6.col-lg-8.course-description > div > div > ul > li")
    topics.append([i.text for i in temp])

    #author_name
    temp = soup.select("#overview > div > div.col-sm-2.col-md-3.col-lg-2.course-meta > div > a > img > cite")
    author_name.append(temp[0].text)

    #author_img
    temp = soup.select("#overview > div > div.col-sm-2.col-md-3.col-lg-2.course-meta > div > a > img")
    author_img.append(temp[0]['data-lazy-src'])

    #tags
    temp = soup.select("#course-tags > div")
    temp1 = temp[0].find_all('a')
    tags.append([i.text for i in temp1])

    #main_img
    temp = soup.select("#banner-thumbnail > img")
    main_img.append(temp[0]['data-lazy-src'])
#######################################################################################
print("downloading author images...")
authors = list(set(zip(author_name, author_img)))
for author in tqdm(authors):
    try:
        file = requests.get(author[1])
    except:
        time.sleep(5)
        file = requests.get(author[1])
    with open(r"D:\\faradars\\files\\lesson8\\authors\\"+author[0]+".jpg", 'wb') as f:
        f.write(file.content)
#######################################################################################
print("downloaing course images...")
for i, img in enumerate(tqdm(main_img)):
    try:
        file = requests.get(img)
    except:
        time.sleep(5)
        file = requests.get(img)
    with open(r"D:\\faradars\\files\\lesson8\\courses\\"+str(i)+".jpg", 'wb') as f:
        f.write(file.content)

#######################################################################################
print("building DataFrame...")
data = pd.DataFrame({
    'title':title,
    'short_description':short_description,
    'duration':duration,
    'skill_level':skill_level,
    'release_date':release_date,

    'description':description,
    'topics':topics,
    'author_name':author_name,
    'author_img':author_img,
    'tags':tags,
    'main_img':main_img,
    
    'link':link
})
data.to_csv(r"D:\faradars\files\lesson8\lynda courses.csv", index=None, encoding='utf-8')
print("finished!")
```
